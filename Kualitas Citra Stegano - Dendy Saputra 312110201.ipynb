```python
# Install the necessary libraries
!pip install Pillow matplotlib

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def calculate_psnr(original, stego):
    # Convert images to NumPy arrays
    original_array = np.array(original)
    stego_array = np.array(stego)

    # Calculate squared error per pixel
    mse = np.mean((original_array - stego_array) ** 2)

    # Calculate PSNR (Peak Signal-to-Noise Ratio)
    if mse == 0:
        return float('inf')
    max_pixel_value = 255.0
    psnr = 20 * np.log10(max_pixel_value / np.sqrt(mse))

    return psnr, mse

def show_images(original, stego):
    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')

    # Display stego image
    plt.subplot(1, 2, 2)
    plt.imshow(stego)
    plt.title('Stego Image')

    plt.show()

def main():
    # Upload original image
    uploaded = files.upload()

    # Get the filename of the original image
    original_image_path = list(uploaded.keys())[0]

    # Upload stego image
    uploaded = files.upload()

    # Get the filename of the stego image
    stego_image_path = list(uploaded.keys())[0]

    # Open original and stego images
    original_image = Image.open(original_image_path)
    stego_image = Image.open(stego_image_path)

    # Display images
    show_images(original_image, stego_image)

    # Calculate PSNR and MSE
    psnr_value, mse_value = calculate_psnr(original_image, stego_image)

    print(f"PSNR between the original and stego images: {psnr_value:.2f} dB")
    print(f"MSE between the original and stego images: {mse_value:.2f}")

if __name__ == "__main__":
    main()
```
